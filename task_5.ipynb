```python
train_df = pd.read_csv(r"C:\Users\Lenovo\Downloads\titanic\train.csv")

test_df = pd.read_csv(r"C:\Users\Lenovo\Downloads\titanic\test.csv")

gender_df = pd.read_csv(r"C:\Users\Lenovo\Downloads\titanic\gender_submission.csv")

print(train_df.head())
```

```python
sns.countplot(x="Survived", data=train_df)
plt.show()

sns.countplot(x="Sex", data=train_df)
plt.show()

sns.countplot(x="Pclass", data=train_df)
plt.show()

sns.countplot(x="Embarked", data=train_df)
plt.show()
```

```python
train_df['Age'].hist(bins=30, edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

sns.boxplot(x=train_df["Age"])
plt.show()

sns.histplot(train_df["Fare"], bins=40, kde=True)
plt.show()
```

```python
sns.countplot(x="Survived", hue="Sex", data=train_df)
plt.show()

sns.countplot(x="Survived", hue="Pclass", data=train_df)
plt.show()

sns.histplot(data=train_df, x="Age", hue="Survived", multiple="stack")
plt.show()

sns.countplot(x="Survived", hue="Embarked", data=train_df)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(train_df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()

sns.pairplot(train_df[["Survived", "Age", "Fare", "Pclass"]], hue="Survived")
plt.show()
```

```python
# Fill missing age with median
train_df["Age"].fillna(train_df["Age"].median(), inplace=True)

# Fill missing embarkation with mode
train_df["Embarked"].fillna(train_df["Embarked"].mode()[0], inplace=True)

# Drop Cabin (too many missing)
train_df.drop(columns=["Cabin"], inplace=True)

train_df.isnull().sum()
```
